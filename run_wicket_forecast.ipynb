```python
# Drop rows where cumulative_wicket is 11
df3 = df3.loc[df3['cumulative_wicket'] != 11]
```

```python
import pandas as pd

# Merge df3 and df2 on 'completed_over' and 'ball_no', but convert float columns to int before merging
df3.loc[:, 'completed_over'] = df3.loc[:, 'completed_over'].astype(int)
df2.loc[:, 'completed_over'] = df2.loc[:, 'completed_over'].astype(int)
df2.loc[:, 'ball_no'] = df2.loc[:, 'ball_no'].astype(int)

# Merge df3 and df2 on 'completed_over' and 'ball_no'
merged_df = df3.merge(df2, on=['completed_over', 'ball_no'], how='left')

# Define a function to calculate the expected_runs based on the conditions
def calculate_expected_runs(row):
    cumulative_wicket = row['cumulative_wicket']
    if cumulative_wicket <= 9:
      wicket_columns = ['zero', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']
      column_name = wicket_columns[cumulative_wicket]
      return row[column_name]
    else:
      return None

# Apply the function to create the 'expected_score' column
merged_df['expected_run'] = merged_df.apply(calculate_expected_runs, axis=1)

# Update df3 with the 'expected_runs' column from merged_df
df3['expected_run'] = merged_df['expected_run']
```

```python
df3['expected_run'] = df3['expected_run'].fillna(method='ffill')
```

```python
# Calculate the difference between 'expected_run' and 'runs_off_bat'
df3['jogi_score'] = df3['expected_run'] - df3['runs_off_bat']
```

```python

# Merge df3 and df4 on 'completed_over', but convert float columns to int before merging
df3.loc[:, 'completed_over'] = df3.loc[:, 'completed_over'].astype(int)
df4.loc[:, 'completed_over'] = df4.loc[:, 'completed_over'].astype(int)

# Merge df3 and df2 on 'completed_over'
merged_df = df3.merge(df4, on=['completed_over'], how='left')

# Define a function to calculate the expected_wicket based on the conditions
def calculate_expected_wicket(row):
    cumulative_wicket = row['cumulative_wicket']
    if cumulative_wicket <= 9:
      wicket_columns = ['zero', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']
      column_name = wicket_columns[cumulative_wicket]
      return row[column_name]
    else:
      return None

# Apply the function to create the 'expected_wicket' column
merged_df['expected_wicket'] = merged_df.apply(calculate_expected_wicket, axis=1)

# Update df3 with the 'jogi_score' column from merged_df
df3['expected_wicket'] = merged_df['expected_wicket']
```

```python
# Calculate the difference between 'expected_wicket' and 'fall_of_wicket'
df3['dismissal_factor'] = df3['fall_of_wicket'] - df3['expected_wicket']
```

```python
for i in range(len(df3['ball_no'])):
  if df3['ball_no'].iloc[i] == 1:
    df3['dismissal_factor'].iloc[i] = df3['dismissal_factor'].fillna(method='bfill').iloc[i]
  else:
    df3['dismissal_factor'].iloc[i] = df3['dismissal_factor'].fillna(method='ffill').iloc[i]
```
